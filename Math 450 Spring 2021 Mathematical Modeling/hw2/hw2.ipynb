(b) Come up with an alternative Fermi model for estimating the earth's radius.

Assume we are given the rotation rate of earth $v_{earth}$ = 460 m/s, and we know earth rotates 24 hours in 1 revolution. Thus we can get the equation: $r_{earth}W_{earth} = v_{earth}$ then we can solve unknown variable $r_{earth} = \frac{W_{earth}}{v_{earth}}$.

$W_{earth} = \frac{2\pi rad}{24hour} = \frac{1hour}{60min} = \frac{1min}{60s}$

Thus,\
$W_{earth} = \frac{2 \pi}{24\times 60\times 60}\quad rad/s$

Then,\
$r_{earth} = \frac{460 \times 24\times 60\times 60}{2 \pi} = \approx 6325454.05824 \approx 6.3 \times 10^{6} m$.

### 4. The calculations Schott actually performs are a little more complicated than described above. Based on other information, some measurements were expected to be more accurate than others. Specifically, the measurement of the angle between Burden to Jescelyne is believed to be less accurate than the other three angles. It would be nice to be able to take this inaccuracy into account when calculating our least-squares solution. And there is indeed away - the method of weighted linear least squares. We can weight the errors produced by each equation such that equations measured to greater accuracy are given larger weights and equations with less accuracy are given smaller weights.

(a) Find the matrix form for the equations of the method of weighted linear least squares by determining the vector $x_{closest}$ that minimizes the total weighted square error

$J(x) = \sum \limits_{t} W_{tt}((Ax)_{t} - b_{t})^{2}$,


where W is a non-negative diagonal matrix.

$
J(x) = \sum \limits_{t} W_{tt}((Ax)_{t} - b_{t})^{2} \\
     = (Ax-b)^{T}W(Ax-b) \\
     = x^{T}A^{T}WAx - x^{T}A^{T}Wb - b^{T}WAx + b^{T}Wb \\
     = x^{T}A^{T}WAx - 2b^{T}WAx + b^{T}Wb 
$

Since $x^{T}A^{T}Wb$ and $b^{T}WAx$ are both scaler then $x^{T}A^{T}Wb = [(x^{T}A^{T})Wb]^{T} = b^{T}WAx$ 

$
\frac{\partial{J}}{x^{T}} = 2x^{T}A^{T}WA - 2b^{T}WA = 0
$

To minimizes the total weighted sqaures error, solve for x:

$x^{T}A^{T}WA = b^{T}WA \\
A^{T}W^{T}Ax_{close} = A^{T}W^{T}b
$

Use the method of linear least squares to find the parabola that best fits the points (-2, 8), (0, 5), (2, 3), (4, 4), and (6, 9).

### 8. A string was wrapped around a metal bar. A 2 ounce weight was hung at one end of the string, and weights W at the other, so as to counterbalance the weight of 2 oz. The more revolutions of string used, the greater an imbalance the friction between string and bar can sustain. The data is in Simple Laws Lecture Notes Exercise #10. Find a law relating the number of revolutions to the maximum sustained weight.

```python
string_data = loadtxt('stringfriction.csv', delimiter=',',comments='#')
string_data
```

```python
A = vstack([string_data[:,0]*0+1, string_data[:,0]]).T
b = log10(string_data[:,1])
est_parms = linalg.solve(A.T.dot(A), A.T.dot(b))
u = A[:,1]
v = est_parms[0] + est_parms[1]*u
plot(u,v,'r-',u,b,'o')
xlabel('N of Revolutions ', fontsize=18)
ylabel('Weight W ($oz$)', fontsize=18)
text(1.0, 0.6, "$W = 10^{%.3f + %.3fn}$"%(est_parms[0], est_parms[1]), fontsize=18)
show()

```

### 9. The chambered nautilus shell is clearly a spiral, but what kind of spiral? The data for this problem is in Simple Laws Lecture Notes Exercise #15.

(a) Fit a logarithmic spiral r = exp(a + $b\theta$) to the data. Find the best-fit values of a and b.
Plot the r$\theta$-curve, the predicted spiral over the observed data, and the radial residuals
of the model as a function of the angle $\theta$.

```python
nau_data = loadtxt('Nautilus3.csv', delimiter=',',comments='#')
nau_data
```

```python
A = vstack([nau_data[:,1]*0+1, nau_data[:,1]]).T
b = log(nau_data[:,0])
est_parms = linalg.solve(A.T.dot(A), A.T.dot(b))
u = A[:,1]
v1 = est_parms[0] + est_parms[1]*u
plot(u,v1,'r-',u,b,'o')
xlabel('angle (radians)', fontsize=18)
ylabel('radius (from image)', fontsize=18)
text(1.5, -1, "$R = e^{%.3f + %.3ft}$"%(est_parms[0], est_parms[1]), fontsize=18)
show()
```

(b) Fit a quadratic polynomial spiral r = p0 + p1$\theta + p_{2}\theta^{2}$ to the data. Find the best-fit
values of p1; p2; p3. Plot the r$\theta$-curve, the predicted spiral over the observed data, and the radial residuals of the model as a function of the angle $\theta$.

```python
A = vstack([nau_data[:,1]*0+1, nau_data[:,1], pow(nau_data[:,1],2)]).T
b = log(nau_data[:,0])
est_parms = linalg.solve(A.T.dot(A), A.T.dot(b))
u = A[:,1]
v2 = est_parms[0] + est_parms[1]*u + est_parms[2]*pow(u,2)
plot(u,v2,'r-',u,b,'o')
xlabel('angle (radians)', fontsize=18)
ylabel('radius (from image)', fontsize=18)
text(1.5, -1, "$r = %.3f + %.3ft + %.3ft^2 }$"%(est_parms[0], est_parms[1], est_parms[2]), fontsize=18)
show()
```
