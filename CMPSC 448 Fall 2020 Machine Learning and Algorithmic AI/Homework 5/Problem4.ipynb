```python
X = [0,0,0,0]
```

```python
p = 0.25
q = 0.25
l = 0.9

diff = 50.0

while diff > pow(10,-3):
    print('test')
    test = X.copy()
    v_0 = l*max(X[1],X[2])
    if X[1] > X[2]:
        print("pick a1")
    else:
        print("pick a2")
    v_1 = l*((1-p)*X[1]+p*X[3])
    v_2 = 1+ l*((1-q)*X[0]+q*X[3])
    v_3 = 10 + l*X[0]
    
    X = [v_0,v_1,v_2,v_3]
    for i in range(len(X)):
        test[i] = abs(test[i]-X[i])
    diff = max(test)
    print(X)
```
