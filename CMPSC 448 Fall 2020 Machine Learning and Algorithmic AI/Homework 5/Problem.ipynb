Carry out a principal component analysis on the entire raw dataset (4-dimensional instances)
for k = 1; 2; 3; 4 components. How much of variance in data can be explained by the first
principal component? How does the fraction of variation explained in data vary as k varies?

```python
var_1st_pc = []
for k in range(1,5):
    pca = PCA(n_components=k)
    #pca.fit(X.copy)
    t_X = pca.fit(X.copy())
    var = pca.explained_variance_ratio_
    print(f"variance (first principal component in k={k}: {var[0]}")
    
    var_1st_pc.append(var[k-1])
    
print(f'list of variance at each eignvalue :{var_1st_pc}')
fraction_var = []
for k in range(1, 5):
    f = var[:k].sum()
    print(f"fraction of variation at k = {k}: {f}")
    fraction_var.append(f)
    
```

```python
fig, axs = plt.subplots()
axs.set_title("result")
axs.set_ylabel("var")
axs.set_xlabel("k")
axs.plot(range(1, 5),fraction_var, label="fraction of variation")
axs.legend()
```

According to the above graph, we can see the fraction of variation explained in data vary as k varies is gradually increasing until reaching k = 4 where k = d = 4.

Apply the preprocessing operations from Problem 1 on raw data set and repeat part (1) on
processed data. Explain any differences you observe compared to part (1) and justify.

```python
X_train = X.copy()
print(X_train.mean(axis=0))
for i in range(4):
    X_train[:,i] = X_train[:,i] - X_train.mean(axis=0)[i]
print(X_train.mean(axis=0))
X_train
```

```python
var_1st_pc = []

mr_scaled_X = X_train
minmax_scaled_X = preprocessing.MinMaxScaler((0,1)).fit_transform(X.copy())
std_scaled_X = preprocessing.StandardScaler().fit_transform(X.copy())
norm_scaled_X = preprocessing.MinMaxScaler().fit_transform(X.copy())

data_pr = [mr_scaled_X,minmax_scaled_X,std_scaled_X,norm_scaled_X]
```

```python
mr_vars = []
minmax_vars = []
std_vars = []
norm_vars = []

var_1st_pc = [mr_vars, minmax_vars, std_vars, norm_vars]

data_pr_name = ['Centering','Scaling of a feature to a range [a,b]','Standardization','Normalization(Between 0 and 1)']

X_list = []

for i in range(len(data_pr)):
    for k in range(1,5):
        print(f"Data Processing: {data_pr_name[i]}")
        pca = PCA(n_components=k)
        t_X = pca.fit_transform(data_pr[i])
        if k == 2:
            X_list.append(t_X)
        var = pca.explained_variance_ratio_
        print(f"variance (first principal component in k={k}: {var[0]}")
        
        var_1st_pc[i].append(var[k-1])

    print(f'list of variance at each eignvalue :{var_1st_pc[i]}')
    fraction_var = []
    for k in range(1, 5):
        f = var[:k].sum()
        print(f"fraction of variation at k = {k}: {f}")
        fraction_var.append(f)
    print('\n\n')
    
    fig, axs = plt.subplots()
    axs.set_title("result:"+data_pr_name[i])
    axs.set_ylabel("var")
    axs.set_xlabel("k")
    axs.plot(range(1, 5),fraction_var, label="fraction of variation")
    axs.legend()
    print('\n\n')

```

Similarly, as we saw before, the differences are the 1st principal component and the increasing tendency of the fraction of variation in the graph. So can see normalization are equivalent to scaling the feature to a range [0,1]. When k = 2, all the processed data have the similar fration of variation which is about 0.95 except clustering since it doesn't include scaling method.

clustering will have the most amount of variance in data can be explained by the first principal component which is about 0.92.
standization will have the least amount of variance in data can be explained by the first principal component which is about 0.72.
